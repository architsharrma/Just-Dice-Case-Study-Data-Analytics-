#### Let's read Ads spend Data:

```python
ads_spend.head()
```

#### 1.Overview of Ad spend data: JustDice produce games and apps,and advertise this in number of different mediums like country and network, which are denoted by country_id and Network_id in ad spend dataset.

#### 2.value_usd: Means how much we spent on different mediums for advertisments 

```python
ads_spend.dtypes
```

```python
ads_spend.shape
```

###### -> From above results, we can see there are 11963 rows and 5 columns

### Insights after revieving Statistical data from above output:
    
##### 1.Max ad spend done in one time is 1152$

##### 2. 75% of ad spend usd value is under 6$ (2022)

##### 3. 21 $ is an average value used for an ad spend in the year 2022.

#### Doing Analysis by taking all columns at once for deep understanding

```python
country_id = ads_spend[['country_id','value_usd']]
```

```python
df = country_id.groupby('country_id').sum()

df.reset_index(inplace=True)

df
```

## insights:

1. maximum dollar has been spend on country id 1
2. minimum dollar has been spend on country id 17

```python
network_id = ads_spend[['network_id','value_usd']]

df12 = network_id.groupby('network_id').sum()

df12.reset_index(inplace=True)

df12
```

### insights:
1.maximum dollar has been spend on network id 60


```python
### Convert event_date into datetime

ads_spend['event_date'] = pd.to_datetime(ads_spend['event_date'])
```

### ->Lets create another column which will be named as month, so that we can map usd_values with months.

```python
ads_spend['Year'] = pd.DatetimeIndex(ads_spend['event_date']).year  
```

```python
ads_spend['Month'] = pd.to_datetime(ads_spend['event_date']).dt.month_name()
```

```python
ads_spend
```

```python
event_date = ads_spend[['Month','value_usd']]

df1 = event_date.groupby('Month').sum().sort_values(by='value_usd',ascending=False)

df1.reset_index(inplace=True)

df1
```

#### Top 3 months where the ad spend value is high is Novermber at the top followed by August and January

## Data Visualization

```python
# fig, ax =plt.subplots(1,2)
plt.figure(figsize=(10.60,4.63))


plt.barh(df1['Month'], df1['value_usd'],linestyle='-',edgecolor='green',color='lightcoral')
plt.title('Total ads spend value correspond to particular months(2022)')
plt.show()
```
